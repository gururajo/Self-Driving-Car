```python
import tensorflow as tf

import numpy as np
import os,pickle,random,cv2
import matplotlib.pyplot as plt
```

```python
path = 'test_set'
tempdata=[]
for img in os.listdir(path+'/cats'):
    if img.endswith('.jpg'):
        tempdata.append([cv2.imread(path+'/cats/'+img),0])
for img in os.listdir(path+'/dogs'):
    if img.endswith('.jpg'):
        tempdata.append([cv2.imread(path+'/dogs/'+img),1])
```

```python
npdata=np.array(tempdata)
```

```python
random.shuffle(tempdata)
```
